```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

recipe_data1 = pd.read_csv("matched_data.csv")

recipe_data1.shape
```

```python
allergy_names = ['milk', 'eggs', 'fish', 'shellfish', 'tree nuts', 'peanuts', 'wheat', 'soy', 'sesame']

allergy_counts = [recipe_data1['milk'].sum(), recipe_data1['eggs'].sum(), recipe_data1['fish'].sum(),
                 recipe_data1['shellfish'].sum(), recipe_data1['tree nuts'].sum(), recipe_data1['peanuts'].sum(),
                 recipe_data1['wheat'].sum(), recipe_data1['soy'].sum(), recipe_data1['sesame'].sum()]

plt.bar(allergy_names, allergy_counts)
plt.ylabel('Number of recipes with allergen')
plt.xticks(rotation='vertical');

# plt.savefig('allergies.png', bbox_inches='tight')
```

```python
for i in range(9):
    print(allergy_names[i], ":", allergy_counts[i])
```

```python
restriction_names = ['no restrictions', 'vegetarian', 'vegan']

no_rest_count = 0
count = 0
for recipe in recipe_data1['name']:
    if (recipe_data1.at[count, 'vegetarian'] == 0) & (recipe_data1.at[count, 'vegan'] == 0):
        no_rest_count += 1
    count += 1

restriction_counts = [no_rest_count, recipe_data1['vegetarian'].sum(), recipe_data1['vegan'].sum()]

plt.bar(restriction_names, restriction_counts)
plt.ylabel('Number of recipes with restriction')

# plt.savefig('restrictions.png', bbox_inches='tight')
```

```python
for i in range(3):
    print(restriction_names[i], ":", restriction_counts[i])
```
