# Optimization of the nutrients

```python
# Trying a new one to print the output #current work
from ortools.sat.python import cp_model


def recipe_optimizer(recipes, targets, Cmin, Cmax, variety_min, variety_max, max_prep, allergens, is_vegan=False, is_vegetarian=False):
    model = cp_model.CpModel()
    sample_size = int(len(recipes) * 0.5) #randomly sample half of the recipes
    sampled_recipes = recipes.sample(n=sample_size, random_state=3) #The random state is set here for consistency however it has been run with various settings

    # Variables
    x = {i: model.NewBoolVar(f'x_{i}') for i in sampled_recipes.index}
    d = {j: model.NewIntVar(0, 100000, f'd_{j}') for j in ['protein_g', 'total_fat_g', 'carbs_g']}

    # Objective: Minimize the sum of absolute deviations from target macronutrients
    model.Minimize(sum(d[j] for j in d))

    # Constraints
    # 1. Macronutrient balance constraints
    for j in ['protein_g', 'total_fat_g', 'carbs_g']:
        nutrient_sum = model.NewIntVar(0, 100000, f'nutrient_sum_{j}')  # Create a variable for the sum
        total_nutrient = sum(x[i] * round(sampled_recipes.loc[i, j]) for i in sampled_recipes.index)
        model.Add(nutrient_sum == total_nutrient)
        model.Add(nutrient_sum - d[j] <= targets[j])
        model.Add(nutrient_sum + d[j] >= targets[j])

    # 2. Dietary and Allergy constraints
    for allergen in allergens:
        if allergen in sampled_recipes.columns:
            for i in sampled_recipes.index:
                if sampled_recipes.loc[i, allergen]:
                    model.Add(x[i] == 0)
    if is_vegan:
        for i in sampled_recipes.index:
            if sampled_recipes.loc[i, 'vegan'] == 0:
                model.Add(x[i] == 0)
    elif is_vegetarian:
        for i in sampled_recipes.index:
            if sampled_recipes.loc[i, 'vegetarian'] == 0:
                model.Add(x[i] == 0)

    # 3. Caloric intake constraints
    caloric_sum = model.NewIntVar(Cmin, Cmax, 'caloric_sum')
    total_calories = sum(x[i] * round(sampled_recipes.loc[i, 'calories']) for i in sampled_recipes.index)
    model.Add(caloric_sum == total_calories)
    model.Add(caloric_sum >= Cmin)
    model.Add(caloric_sum <= Cmax)

    # 4. Meal diversity and quantity constraints
    model.Add(sum(x[i] for i in sampled_recipes.index) >= variety_min)
    model.Add(sum(x[i] for i in sampled_recipes.index) <= variety_max)

    # 5. Maximum time constraint
    time_sum = model.NewIntVar(0, max_prep, 'time_sum')
    total_time = sum(x[i] * round(sampled_recipes.loc[i, 'minutes']) for i in sampled_recipes.index)
    model.Add(time_sum == total_time)
    model.Add(time_sum <= max_prep)

    # Solver configuration and solution
    solver = cp_model.CpSolver()
    status = solver.Solve(model)

    selected_recipes = []
    if status == cp_model.OPTIMAL:
        selected_recipes = [i for i in sampled_recipes.index if solver.Value(x[i])]
        print("Selected recipes indices:", selected_recipes)
    else:
        print("No optimal solution found.")
    
    return selected_recipes, sampled_recipes
```

```python

Cmin = int(period_calories * 0.95) # Minimum calorie variation 5%
Cmax = int(period_calories * 1.05)  # Maximum 5% above target
variety_min = num_meals  # Example minimum number of recipes
variety_max = num_meals  # Example maximum number of recipes

#Getting allergens from input
# Creating a dictionary to map checkbox states to allergen names
allergens_dict = {
    'milk': milk_allergy,
    'eggs': eggs_allergy,
    'fish': fish_allergy,
    'shellfish': shellfish_allergy,
    'tree nuts': treenuts_allergy,
    'peanuts': peanuts_allergy,
    'wheat': wheat_allergy,
    'soy': soy_allergy,
    'sesame': sesame_allergy
}

# Filtering to get only those allergens where the checkbox is True
allergens = [allergen for allergen, is_allergic in allergens_dict.items() if is_allergic]

#vegan and vegetarian initialized already above checking sstatus
if is_vegetarian:
    print("Dietary restriction: Vegetarian")
if is_vegan:
    print("Dietary restriction: Vegan")


# Now, selected_allergens can be passed to the optimizer function
print("Allergens to be included in the optimizer:", allergens)

# Targets for macronutrients
targets = {
    'protein_g': period_protein,  # Target grams of protein
    'total_fat_g': period_fat,  # Target grams of fat
    'carbs_g': period_carbs  # Target grams of carbohydrates
}
```

```python
# Calling the function
start_time = time.time()
selected_indices, sampled_recipes = recipe_optimizer(recipe_data1, targets, Cmin, Cmax, variety_min, variety_max, max_prep, allergens, is_vegan, is_vegetarian)
end_time = time.time()
runtime = end_time - start_time
print(f"Runtime ortools: {runtime} seconds")
```

```python
def analyze_deviations(recipes, selected_indices, targets):
    selected_recipes = recipes.loc[selected_indices]

    # Calculate actual sums
    actual_totals = {
        'protein_g': selected_recipes['protein_g'].sum(),
        'total_fat_g': selected_recipes['total_fat_g'].sum(),
        'carbs_g': selected_recipes['carbs_g'].sum()
    }

    # Calculate deviations
    deviations = {nutrient: actual_totals[nutrient] - targets[nutrient] for nutrient in targets}

    # Print results
    print("Analysis of Deviations:")
    print(f"Selected recipes indices: {selected_indices}")
    print(f"Total protein: {actual_totals['protein_g']} g (Target: {targets['protein_g']} g, Deviation: {deviations['protein_g']} g)")
    print(f"Total fat: {actual_totals['total_fat_g']} g (Target: {targets['total_fat_g']} g, Deviation: {deviations['total_fat_g']} g)")
    print(f"Total carbs: {actual_totals['carbs_g']} g (Target: {targets['carbs_g']} g, Deviation: {deviations['carbs_g']} g)")

    return deviations
```
