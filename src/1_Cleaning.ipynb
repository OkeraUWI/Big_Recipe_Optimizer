```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('RAW_recipes.csv')

df.info()
```

```python
df.head()
```

```python
#Removing unnecessary columns

df.drop('id', axis=1, inplace=True)
df.drop('contributor_id', axis=1, inplace=True)
df.drop('submitted', axis=1, inplace=True)
df.drop('n_steps', axis=1, inplace=True)
df.drop('steps', axis=1, inplace=True)
df.drop('n_ingredients', axis=1, inplace=True)

df.head()
```

```python
#finding unique tags

all_tags = []

for i in df['tags']:
    y = i.split("'")
    for j in y:
        if len(j) > 2:
            all_tags.append(j)
            
print(np.unique(all_tags))
```

```python
#convert tags, nutrition and ingredients from strings to arrays

count = 0
for i in df['tags']:
    tags_list = []
    y = i.split("'")
    for j in y:
        if len(j) > 2:
            tags_list.append(j)
    df.at[count, 'tags'] = tags_list
    count += 1

count = 0    
for i in df['ingredients']:
    ing_list = []
    y = i.split("'")
    for j in y:
        if len(j) > 2:
            ing_list.append(j)
    df.at[count, 'ingredients'] = ing_list
    count += 1

count = 0
for i in df['nutrition']:
    j = i.strip('[]')
    df.at[count,'nutrition'] = [float(num) for num in j.split(',')]
    count += 1
```

```python
#categorize as vegetarian, vegan

df['vegan'] = 0
df['vegetarian'] = 0

count = 0
for list in df['tags']:
    for tag in list:
        if tag == 'vegan':
            df.at[count, 'vegan'] = 1
    count += 1
    
count = 0
for list in df['tags']:
    for tag in list:
        if tag == 'vegetarian':
            df.at[count, 'vegetarian'] = 1
    count += 1
```

```python
#drop na values
df.dropna(inplace=True)
df.reset_index(drop=True, inplace=True)
df.shape
```

```python
# Delete unhealthy/irrelevant recipes

import re

bad_tags = ['cakes', 'chocolate', 'snacks', 'bar-cookies', 'candy', 'cheesecake', 'cake-fillings-and-frostings', 'coffee-cakes', 'brownies',
            'beverages', 'chocolate-chip-cookies', 'cocktails', 'cookies-and-brownies', 'cranberry-sauce',
            'desserts', 'desserts-easy', 'desserts-fruit', 'fillings-and-frostings-chocolate', 'frozen-desserts',
            'halloween-cakes', 'halloween-cocktails', 'halloween-cupcakes', 'jams-and-preserves',
            'jellies', 'puddings-and-mousses', 'rolled-cookies', 'snacks', 'snacks-kid-friendly', 'snacks-sweet',
            'sweet-sauces', 'dip', 'sauce', 'spice mix', 'ketchup']

count = 0
for list in df['tags']:
    for tag in list:
        if tag in bad_tags:
            df.at[count,'tags'] = float('nan')
    count += 1
    
count = 0
for n in df['name']:
    for tag in bad_tags:
        if re.search(f".*{tag}.*", n):
            df.at[count,'name'] = float('nan')
    count += 1

df.dropna(inplace=True)
df.reset_index(drop=True, inplace=True)
df.shape
```

```python
allergies = {'milk': ['milk', 'yogurt', 'butter', 'cheese', 'cream', 'ghee', 'whey' , 'caramel', 'chocolate' 'margarine',
                      'lactose'], 
             'eggs': ['egg', 'mayo', 'meringue'], 
             'fish': ['fish', 'anchovies', 'bass', 'catfish', 'cod', 'flounder', 'grouper', 'haddock', 'hake', 'halibut',
                      'herring', 'mahi', 'perch', 'pike', 'pollock', 'salmon', 'scrod', 'sole', 'snapper',
                      'tilapia', 'trout', 'tuna', 'barbecue', 'worcestershire'],
             'shellfish': ['shellfish', 'barnacle', 'crab', 'crawfish', 'crawdad', 'ecrevisse', 'fish', 'krill', 
                           'lobster', 'scampi', 'prawns', 'shrimp', 'abalone', 'clams', 'cockle', 'limpet', 'mussels', 
                           'octopus', 'oysters', 'periwinkle', 'sea cucumber', 'sea urchin' 'scallops', 'snails', 'squid', 
                           'calamari', 'whelk', 'seafood'],
             'tree nuts': ['nut', 'almond', 'cashew', 'gianduja', 'pesto', 'pecan', 'pistachio'], 
             'peanuts': ['peanut'], 
             'wheat': ['wheat', 'flour', 'bread', 'pasta', 'noodle', 'semolina', 'couscous', 'orzo'], 
             'soy': ['soy', 'edamame', 'tofu', 'miso'], 
             'sesame': ['sesame', 'tahini', 'benne', 'benniseed' , 'gingelly' ,'gomasio']
            }

for allergy in allergies:
    df[allergy] = 0
    count = 0
    for list in df['ingredients']:
        if type(list) == float:
            print(count)
        for ingredient in list:
            for allergen in allergies[allergy]:
                if re.search(f".*{allergen}.*", ingredient):
                    df.at[count, allergy] = 1     
        count += 1
    count1 = 0
    for list in df['tags']:
        for tag in list:
            for allergen in allergies[allergy]:
                if re.search(f".*{allergen}.*", tag):
                    df.at[count1, allergy] = 1               
        count1 += 1

df.head()    
```

```python
df['calories'] = df.apply(lambda x: x['nutrition'][0], axis=1)
df['total_fat_pdv'] = df.apply(lambda x: x['nutrition'][1], axis=1)
df['sugar_pdv'] = df.apply(lambda x: x['nutrition'][2], axis=1)
df['sodium_pdv'] = df.apply(lambda x: x['nutrition'][3], axis=1)
df['protein_pdv'] = df.apply(lambda x: x['nutrition'][4], axis=1)
df['saturated_fat_pdv'] = df.apply(lambda x: x['nutrition'][5], axis=1)
df['carbs_pdv'] = df.apply(lambda x: x['nutrition'][6], axis=1)

df.head()
```

```python
#tag each as breakfast, lunch, dinner

df['breakfast'] = 0
df['lunch'] = 0
df['dinner'] = 0

count = 0
for list in df['tags']:
    for tag in list:
        if (tag == 'breakfast') or (re.search(f".*breakfast.*", df.at[count,'description'])) or (re.search(f".*breakfast.*", df.at[count,'name'])):
            df.at[count,'breakfast'] = 1
        if (tag == 'lunch') or (re.search(f".*lunch.*", df.at[count,'description'])) or (re.search(f".*lunch.*", df.at[count,'name'])):
            df.at[count,'lunch'] = 1
        if (tag == 'dinner') or (re.search(f".*dinner.*", df.at[count,'description'])) or (re.search(f".*dinner.*", df.at[count,'name'])):
            df.at[count,'dinner'] = 1
    count += 1

df.head()
```

```python
df.drop('tags', axis=1, inplace=True)
df.drop('description', axis=1, inplace=True)
df.head()
```
